```python
# Import the numpy, pandas ,seaborn and matplotlib packages

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import csv
#importing csv file from web
uber =pd.read_csv('https://cdn.upgrad.com/UpGrad/temp/76b3b6a4-d87d-4e82-b1c3-3f6e10b9c076/Uber%20Request%20Data.csv')
uber
```

```python
# Inspect the dataframe
print(uber.shape)

uber.info()

uber.describe()
```

```python
# Checking for duplicated rows
sum(uber.duplicated(subset='Request id'))
```

```python
# checking for column-wise null count here
print(uber.isnull().sum())
```

```python
#  code for column-wise null percentages here
round(100*(uber.isnull().sum()/len(uber.index)),2)
```

```python
#Analyzing the NA values in Driver_id column:
uber[uber['Driver id'].isnull()]

#For the "Driver_id" column the values are NA for only those "Request_id" where the "Status" is "No Cars Available"
#At this time there is no need to remove the NA values from the Driver_id column.
```

```python
#  code for filling the NaN values in the 'Driver id' column here
uber['Driver id'].fillna('None',inplace=True)

round(100*(uber.isnull().sum()/len(uber.index)),2)
```

```python
#Analyzing the NA values in Drop timestamp column:
uber[uber['Drop timestamp'].isnull()]
#For the "Drop_timestamp" column the values are NA for only those "Request_id" where the "Status" is "No Cars Available"
#or "Cancelled"
##At this time there is no need to remove the NA values from the Driver_id column.
```

```python
# convert Request timestamp &  Drop timestamp to date-time object
uber['Request timestamp'] = pd.to_datetime(uber['Request timestamp'],dayfirst=True)
uber['Drop timestamp'] = pd.to_datetime(uber['Drop timestamp'],dayfirst=True)

uber.info()
```

```python
#  code for filling the NaN values in the 'Drop timestamp' column here
uber['Drop timestamp'].fillna('None',inplace=True)

round(100*(uber.isnull().sum()/len(uber.index)),2)
```

```python
# Extract hour and day from request timestamp
uber['Request_hour'] = uber['Request timestamp'].dt.hour

uber['day_of_week'] = uber['Request timestamp'].dt.day_name()
uber
```

```python
# Deriving Time_slot column by Dividing hours of day.
uber['time_slot'] = pd.cut(uber['Request_hour'], [0,4,10,14,20,24], labels=["Late Night", "Morning","Day_Time","Evening","Night"],right=False)

uber.time_slot.value_counts()
```

```python
#Univariate Analusis
#Analysis of pickup ponit
#Total Request by pickup location
fig_Pickup=uber['Pickup point'].value_counts().plot(kind='bar')
fig_Pickup.set_title('Pickup point Analysis')
fig_Pickup.set_xlabel('Pickup point')
fig_Pickup.set_ylabel('No. of Count')
```

```python
#Analysis of Status
#number of request Status
fig_status=uber['Status'].value_counts().plot(kind='bar')
fig_status.set_title('Status Analysis')
fig_status.set_xlabel('Status')
fig_status.set_ylabel('No. of Count')
```

```python
# Aggregate count plot for all days hours w.r.t to Pickup point
sns.catplot(x = "Request_hour", hue = "Pickup point", data = uber, kind = "count")
```

```python
#Analysing pickup point and status w.r.t time slot
plt.figure(figsize=(30,20))
uber.groupby(['Pickup point','Status','time_slot']).size().unstack(fill_value=0).plot(kind='bar')
```

```python
#Analysing pickup point and days of week w.r.t status
plt.figure(figsize=(30,20))
uber.groupby(['Pickup point','day_of_week','Status']).size().unstack(fill_value=0).plot(kind='bar')
```

```python
# plot time slot w.r.t. status and identify the most critical problems 
sns.countplot(x = "time_slot", hue = "Status", data = uber)
```

```python
# in the above plot the higest number of cancellations are in the "Morning" time slot
#find the pickup point fron where most cancellations occurs
uber_morning_Status = uber[uber.time_slot == "Morning"]
sns.countplot(x="Pickup point", hue="Status", data = uber_morning_Status)
```

```python
# in the above plot the higest number of no cars availabl are in the "Evening" time slot
#find the pickup point fron where most occurence of no cars available
uber_eve_status = uber[uber.time_slot == "Evening"]
sns.countplot(x="Pickup point", hue="Status", data = uber_eve_status)
```

```python
#finding Total demand wrt time slot and Pickup point
group_by_time_Pickup = uber.groupby(['time_slot','Pickup point'])
total_Demand=pd.DataFrame(group_by_time_Pickup['Request id'].count().sort_values(ascending=False))
total_Demand
```

```python
#finding Total supply (i.e trips completed ) wrt Time slot and pickup point
Trip_completed = pd.DataFrame(uber.loc[uber['Status']=="Trip Completed"])
group_by_time_Pickup_status = Trip_completed.groupby(['time_slot','Pickup point'])
total_Supply=pd.DataFrame(group_by_time_Pickup_status['Request id'].count().sort_values(ascending=False))
total_Supply
```

```python
#adding supply(total no of request completed) and demand(total request arise) columns w.r t time slot in  uber dataframe 
uber["Demand"] = uber.groupby(['time_slot','Pickup point'])["Request id"].transform('count')
uber["Supply"] = Trip_completed.groupby(['time_slot','Pickup point'])["Request id"].transform('count')
uber
```

```python
#plotting the gap between supply and demand w.r.t Pickup point


sns.barplot(x="Demand", y="Supply", hue="Pickup point", data=uber)

```
